```python
from pydub import AudioSegment
import os

def convert_mp3_to_wav(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    for file in os.listdir(input_folder):
        if file.endswith(".mp3"):
            mp3_path = os.path.join(input_folder, file)
            wav_path = os.path.join(output_folder, os.path.splitext(file)[0] + ".wav")
            
            audio = AudioSegment.from_mp3(mp3_path)
            audio.export(wav_path, format="wav")
            print(f"Converted: {file} -> {wav_path}")

# Example usage
input_folder = r"C:\Users\Kanishk\Downloads\bird_sounds\mp3"  # Folder containing MP3 files
output_folder = r"C:\Users\Kanishk\Downloads\bird_sounds\wav"
convert_mp3_to_wav(input_folder, output_folder)
```
